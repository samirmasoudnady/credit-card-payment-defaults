### Dataset Link : https://www.kaggle.com/datasets/uciml/default-of-credit-card-clients-dataset

- ID: ID of each client
- LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit
- SEX: Gender (1=male, 2=female)
- EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
- MARRIAGE: Marital status (1=married, 2=single, 3=others)
- AGE: Age in years
- PAY_0: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, … 8=payment delay for eight months, 9=payment delay for - nine months and above)
- PAY_2: Repayment status in August, 2005 (scale same as above)
- PAY_3: Repayment status in July, 2005 (scale same as above)
- PAY_4: Repayment status in June, 2005 (scale same as above)
- PAY_5: Repayment status in May, 2005 (scale same as above)
- PAY_6: Repayment status in April, 2005 (scale same as above)
- BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)
- BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)
- BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)
- BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)
- BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)
- BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)
- PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)
- PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)
- PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)
- PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)
- PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)
- PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)
- default.payment.next.month: Default payment (1=yes, 0=no)

# Data Cleaning

```python
# fun to replace all value -2 and -1 with 0 "No delay"
def clean_pay_cols(x):

    if x in [-2, -1]:
        return 0

    else:
        return x

df[['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4',	'PAY_5', 'PAY_6']] = df[['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4',	'PAY_5', 'PAY_6']].applymap(clean_pay_cols)
df[['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4',	'PAY_5', 'PAY_6']]
```

```python
df.columns = df.columns.str.strip()
df.columns
```


#### EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others ,5=unknown, 6=unknown)


```python
def clean_edu(x):

    if x in ['Unknown', '0']:
        return 'Others'

    else:
        return x

df['EDUCATION'] = df['EDUCATION'].apply(clean_edu)
df['EDUCATION']
```

#### MARRIAGE: Marital status (1=married, 2=single, 3=others)

```python
def clean_marriage(x):

    if x == '0':
        return 'Other'

    else:
        return x

df['MARRIAGE'] = df['MARRIAGE'].apply(clean_marriage)
df['MARRIAGE']
```

### Data Preprocessing

### Split Data into Input features and Target variable

```python
x = df.drop(['default.payment.next.month'], axis= 1)
y = df['default.payment.next.month']
```

### Split Data into Train & Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42, stratify= y)
```

### Handle Numerical columns

```python
scaling_cols = x_train.drop(['SEX', 'EDUCATION', 'MARRIAGE','PAY_0', 'PAY_2', 'PAY_3', 'PAY_4',	'PAY_5', 'PAY_6'], axis= 1).columns
scaling_cols
```

```python
from sklearn.preprocessing import RobustScaler

rc = RobustScaler()

x_train[scaling_cols] = rc.fit_transform(x_train[scaling_cols])
x_test[scaling_cols] = rc.transform(x_test[scaling_cols])
```

### Handle Categorical Columns

```python
ohe_cols = ['SEX', 'MARRIAGE']
ohe_cols
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop= 'first', sparse_output= False)

ohe_train_arr = ohe.fit_transform(x_train[ohe_cols])
ohe_test_arr = ohe.transform(x_test[ohe_cols])
```

```python
ohe_train_df = pd.DataFrame(ohe_train_arr, columns= ohe.get_feature_names_out())
ohe_test_df = pd.DataFrame(ohe_test_arr, columns= ohe.get_feature_names_out())
```

```python
ohe_test_df
```

```python
x_train.reset_index(inplace= True, drop= True)
x_test.reset_index(inplace= True, drop= True)

y_train.reset_index(inplace= True, drop= True)
y_test.reset_index(inplace= True, drop= True)
```

```python
x_train = pd.concat([x_train, ohe_train_df], axis= 1).drop(ohe_cols, axis= 1)

x_test = pd.concat([x_test, ohe_test_df], axis= 1).drop(ohe_cols, axis= 1)
```

### Machine Learning

### 1) use Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

lr = LogisticRegression(random_state=42, n_jobs=-1)

lr.fit(x_train, y_train)

print('Train LR :\n', classification_report(y_train, lr.predict(x_train)))

print('\nTest LR \n:', classification_report(y_test, lr.predict(x_test)))
```

### 1.1) use Logistic Regression model with class weight

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

lr = LogisticRegression(random_state=42, n_jobs=-1, class_weight='balanced')

lr.fit(x_train, y_train)

print('Train LR :\n', classification_report(y_train, lr.predict(x_train)))

print('\nTest LR \n:', classification_report(y_test, lr.predict(x_test)))
```

### 2) KNN classifier 7 neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors= 7, n_jobs=-1)

knn.fit(x_train, y_train)

print('Train knn :\n', classification_report(y_train, knn.predict(x_train)))

print('\nTest knn \n:', classification_report(y_test, knn.predict(x_test)))
```

### 3) SVM with class weight

```python
from sklearn.svm import SVC

svc = SVC(class_weight='balanced')

svc.fit(x_train, y_train)

print('Train svc :\n', classification_report(y_train, svc.predict(x_train)))

print('\nTest svc \n:', classification_report(y_test, svc.predict(x_test)))
```

### 4) Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

gaussuan_nb = GaussianNB()

gaussuan_nb.fit(x_train, y_train)

print('Train gaussuan_nb :\n', classification_report(y_train, gaussuan_nb.predict(x_train)))

print('\nTest gaussuan_nb \n:', classification_report(y_test, gaussuan_nb.predict(x_test)))
```

### 5) XGB

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estmotors= 500, max_depth=5, learning_rate=0.05, n_jobs=-1, random_state=42)

xgb.fit(x_train, y_train)

print('Train xgb :\n', classification_report(y_train, xgb.predict(x_train)))

print('\nTest xgb \n:', classification_report(y_test, xgb.predict(x_test)))
```

### 6) CAT BOOST

```python
from catboost import CatBoostClassifier

cat = CatBoostClassifier(iterations=500, learning_rate=0.1, depth=6, verbose=0)

cat.fit(x_train, y_train)

print('Train cat boost :\n', classification_report(y_train, cat.predict(x_train)))

print('\nTest cat boost \n:', classification_report(y_test, cat.predict(x_test)))
```

### 6.1) Cat Boost with class weight

```python
from catboost import CatBoostClassifier

cat = CatBoostClassifier(iterations=500, learning_rate=0.1, depth=6, scale_pos_weight = len(y_train[y_train==0]) / len(y_train[y_train==1]), verbose=0)

cat.fit(x_train, y_train)

print('Train cat boost :\n', classification_report(y_train, cat.predict(x_train)))

print('\nTest cat boost \n:', classification_report(y_test, cat.predict(x_test)))
```

### 6.2) Cat Boost with (SMOTE)

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)

cat = CatBoostClassifier(iterations=300, learning_rate=0.1, scale_pos_weight = len(y_train[y_train==0]) / len(y_train[y_train==1]), verbose=0)

cat.fit(x_train_smote, y_train_smote)

print('Train xgb :\n', classification_report(y_train_smote, cat.predict(x_train_smote)))

print('\nTest xgb \n:', classification_report(y_test, cat.predict(x_test)))
```

### 6.2) Cat Boost with (over sampling teq)

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
x_train_ros, y_train_ros = ros.fit_resample(x_train, y_train)

cat = CatBoostClassifier(iterations=500, learning_rate=0.1, verbose=0)

cat.fit(x_train_smote, y_train_smote)

print('Train cat boost :\n', classification_report(y_train_ros, cat.predict(x_train_ros)))

print('\nTest cat boost \n:', classification_report(y_test, cat.predict(x_test)))
```

### 6.3) Cat Boost with (over_sampling & class weight & threshold) it is the best model

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
x_train_ros, y_train_ros = ros.fit_resample(x_train, y_train)

cat = CatBoostClassifier(iterations=500, learning_rate=0.01, scale_pos_weight = len(y_train[y_train==0]) / len(y_train[y_train==1]), verbose=0)

cat.fit(x_train_smote, y_train_smote)
# predict propabilty
y_pred_proba = cat.predict_proba(x_test)[:,1]

# apply threshhold manuly
threshold = 0.8
y_pred = (y_pred_proba >= threshold).astype(int)

print('Train cat boost :\n', classification_report(y_train_ros, cat.predict(x_train_ros)))
print('Test cat boost :\n', classification_report(y_test, y_pred))
```

```python
# to save model
import joblib
joblib.dump({"model": cat, "threshold": threshold}, "catboost_with_threshold.pkl")
```
